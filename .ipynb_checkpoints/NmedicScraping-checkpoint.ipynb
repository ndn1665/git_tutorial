```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import quote
import time
import json
import requests
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
pre_json_diseases = dict()
letter_list = ['가','나','다','라','마','바','사','아','자','차','카','타','파','하','A']
page_list = [20,5,6,3,7,10,17,20,13,6,2,3,5,8,2]
for letter in letterList:
    quoted_letter = quote(letter)    
    for i in range(1,21):
        url = 'https://www.snuh.org/health/nMedInfo/nList.do?pageIndex='+str(i)+'&searchNWord=' + quoted_letter
        web = requests.get(url, headers=headers,timeout = 10).content
        web_page = BeautifulSoup(web, 'html.parser')

        disease_names = web_page.find_all('div',{'class':'item'})
        for pre_disease_name in disease_names:
            a_tag = pre_disease_name.find('div',{'class': 'title'}).find('a')
            if a_tag:
                link = a_tag.get('href')[2:]
                disease_name = pre_disease_name.find('div',{'class': 'title'}).get_text()
                tmp_name = ''
                disease_name = disease_name.split('[')[0].strip('\n').strip(' ').split(' ')
                for i in disease_name:
                    tmp_name += i
                disease_name = tmp_name
                info_url = 'https://www.snuh.org/health/nMedInfo/'+link
                info_web = requests.get(info_url, headers=headers,timeout = 10).content
                info_web_page = BeautifulSoup(info_web, 'html.parser')
    
                hospitals = info_web_page.find('div',{'class':'viewRow tooltipRow'}).find_all('a')
                pre_json_diseases[disease_name] = {"count":0,"hospitals":[]}
                count = 0
                for hospital in hospitals:
                    count +=1
                    pre_json_diseases[disease_name]["hospitals"].append(hospital.get_text())
            
                pre_json_diseases[disease_name]["count"] = count
        time.sleep(0.5)
    time.sleep(10)

with open('json_diseases.json', 'w', encoding='utf-8') as make_file:

    json.dump(pre_json_diseases, make_file, indent="\t",ensure_ascii = False)
```

```python


pre_json_diseases = dict()

url = 'https://www.snuh.org/health/nMedInfo/nList.do?pageIndex=1&searchNWord=' + quote('가')
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
web = requests.get(url, headers=headers).content
web_page = BeautifulSoup(web, 'html.parser')


disease_names = web_page.find_all('div',{'class':'item'})
for pre_disease_name in disease_names:
    a_tag = pre_disease_name.find('div',{'class': 'title'}).find('a')
    if a_tag:
        link = a_tag.get('href')[2:]
        disease_name = pre_disease_name.find('div',{'class': 'title'}).get_text()
        tmp_name = ''
        disease_name = disease_name.split('[')[0].strip('\n').strip(' ').split(' ')
        for i in disease_name:
            tmp_name += i
        disease_name = tmp_name
        info_url = 'https://www.snuh.org/health/nMedInfo/'+link
        info_web = urlopen(info_url).read().decode('utf-8')
        info_web_page = BeautifulSoup(info_web, 'html.parser')

        hospitals = info_web_page.find('div',{'class':'viewRow tooltipRow'}).find_all('a')
        
        pre_json_diseases[disease_name] = {"count":0,"hospitals":[]}
        count = 0
        for hospital in hospitals:
            count +=1
            pre_json_diseases[disease_name]["hospitals"].append(hospital.get_text())
            
        pre_json_diseases[disease_name]["count"] = count
        
    time.sleep(0.1)
with open('test2.json', 'w', encoding='utf-8') as make_file:

    json.dump(pre_json_diseases, make_file, indent="\t",ensure_ascii=False)

# 저장한 파일 출력하기

with open('test2.json', 'r',encoding='utf-8') as f:

    json_data = json.load(f)

print(json.dumps(json_data, indent="\t",ensure_ascii=False) )
```
